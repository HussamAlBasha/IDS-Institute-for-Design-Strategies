# <h1 style="color:#AED6F1 ; font-weight: bold; font-style: italic;">The Soundscape Shape of All Locations in the Data Sheet</h1>

```python
df = df.sspy.add_paq_coords()
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
for i, location in enumerate(df.LocationID.unique()):
    df.isd.filter_location_ids(location_ids=[location]).sspy.density(
        ax=axes.flatten()[i], title=location
    )

plt.tight_layout()
```
